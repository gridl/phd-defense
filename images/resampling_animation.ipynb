```python
chosen = [21, 35, 15, 24, 29, 26, 29, 17, 27]
```

```python
cummean = np.cumsum(chosen) / np.arange(1, len(chosen)+1)
cummean
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axhline(y=25, color='C1')
ax.set_ylim(20.5, 28.5)
ax.set_xlim(0.5, 9.5)
x = np.arange(1, len(cummean)+1)
ax.plot(x[:i], cummean[:i], '-o')
```

```python
for i in range(1, len(chosen)+1):
    fig, ax = plt.subplots()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.axhline(y=25, color='C1')
    ax.set_ylim(20.5, 28.5)
    ax.set_xlim(0.5, 9.5)
    x = np.arange(1, len(cummean)+1)
    ax.plot(x[:i], cummean[:i], '-o')
    fig.savefig("resampling/step-{0}.svg".format(i))
```
