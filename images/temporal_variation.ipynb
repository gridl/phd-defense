```python
data = pd.read_csv("VMMdata.csv", index_col='Time', parse_dates=True)
```

```python
data = data.resample('1min').mean()
```

```python
fig, ax = plt.subplots()
data['BCR802'].plot(ax=ax)
ax.set_xlim("2012-02-17 10:00", "2012-02-17 15:00")
ax.set_ylim(0, 15)
ax.set_ylabel("Black carbon (Âµg/mÂ³)")
ax.grid()
fig.savefig("temporal-variation-1b.png")
```

```python
fig, ax = plt.subplots()
data['BCR802'].resample('D').mean().plot(ax=ax)
ax.set_ylim(0, 8)
ax.set_ylabel("Black carbon (Âµg/mÂ³)")
ax.grid()
ax.set_xlabel("Time (days)")
fig.savefig("temporal-variation-3b.png")
```

```python
fig, ax = plt.subplots()
data['BCR802'].groupby(data.index.hour).mean().plot(ax=ax)
ax.set_ylim(0, 8)
ax.set_ylabel("Black carbon (Âµg/mÂ³)")
ax.grid()
ax.set_xlabel("Hour of the day (h)")
fig.savefig("temporal-variation-2b.png")
```
